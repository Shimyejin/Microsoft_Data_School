```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://news.sbs.co.kr/news/Subusu_RssFeed.do?plink=RSSREADER'
response = requests.get(url)

# -------------------
# xml 파서 이용
# -------------------
soup = BeautifulSoup(response.text, 'xml')

# -------------------------------------
# 기사제목
# -------------------------------------
titles = soup.select('item>title')
titles = [title.text for title in titles]

# -------------------------------------
# 기사링크로 이동하여 컨텐츠 수집
# -------------------------------------
links = soup.select('item>link')
contents = []
for link in links:
    response = requests.get(link.text)
    soup = BeautifulSoup(response.text, 'html.parser')
    content = soup.select_one('.main_text>.text_area').text
    contents.append(content)


```

```python
# -------------------------------------
# 데이터프레임 생성
# -------------------------------------
df = pd.concat([pd.Series(titles), pd.Series(contents)], axis=1)
df.columns = ['title', 'content']
df['content'].iloc[0]
```
