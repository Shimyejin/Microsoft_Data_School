```python
import pandas as pd
```

### 기상 데이터프레임 생성

* 여러 데이터파일을 연결하여 하나의 데이터프레임 생성
* 데이터 파일 리스트 생성
    * `import glob`
    * `glob.glob(파일경로규칙)`

```python
# ---------------------------
# 데이터 파일 리스트 생성
# ---------------------------

import glob

# data/폴더에서 '기상데이터'가 파일명에 포함된 모든 csv 파일
# 파일명 앞에는 한글자 이상의 다른 글자가 포함되어야 함

files = glob.glob("C:\\Users\\User\\Documents\\VSCode\\Data School\\data school\\00.data\\기상데이터\\*기상데이터.csv")
files
```

```python
# ---------------------------
# 데이터프레임 리스트 생성
# ---------------------------

dfs = []  # 빈 리스트 생성

for file in files:
    df = pd.read_csv(file)  # 혹은 encoding='utf-8-sig'로 변경할 수도 있음
    dfs.append(df)
```

```python
# ---------------------------
# 데이터프레임 리스트의 길이 확인
# ---------------------------

len(dfs)

# ---------------------------
# 데이터프레임 리스트의 개별 데이터프레임 확인
# ---------------------------

dfs[0].head()
```

```python
# ---------------------------
# 데이터프레임 연결
#   df_all
# ---------------------------

df_all = pd.concat(dfs, ignore_index=True)
```

```python
# ---------------------------
# 연결한 데이터프레임 shape 확인
# ---------------------------

print("shape:", df_all.shape)

# ---------------------------
# 연결한 데이터프레임 head, tail 확인
# ---------------------------

print("head()>>>")
df_all.head()
```

```python
print("tail()>>>")
df_all.tail()
```

```python
# ---------------------------
# info()
# ---------------------------

df_all.info()
```

```python
# ---------------------------
# 결측치 확인
# ---------------------------

df_all.isnull().sum()
```

```python
# ---------------------------
# 연도, 월 컬럼 추가
# ---------------------------

df_all['일시'] = pd.to_datetime(df_all['일시'])  # 문자열을 datetime으로 변환
df_all['연도'] = df_all['일시'].dt.year
df_all['월'] = df_all['일시'].dt.month
```

### 지점정보 데이터프레임 생성
* 지점, 지점명, 데이터수, 시작년도, 끝년도

```python
# --------------------
# 지점별 데이터 수
#   df_location_cnt
# --------------------

df_location_cnt = df_all.groupby(['지점', '지점명'])['일시'].count().to_frame(name='데이터수')
df_location_cnt
```

```python
# -------------------------
# 지점별 시작년도, 끝년도
#   df_location_year
# -------------------------

df_location_year = df_all.groupby(['지점', '지점명'])['연도'].agg(
    시작년도='min',
    끝년도='max'
)
df_location_year
```

```python
df_location = pd.concat([df_location_cnt, df_location_year], axis=1)
df_location
```

```python
# -------------------------
# 지점별 데이터 수, 시작년도, 끝년도 concat
#   df_location
# -------------------------

df_location = pd.concat([df_location_cnt, df_location_year], axis=1)
df_location = df_location.reset_index()
df_location
```
