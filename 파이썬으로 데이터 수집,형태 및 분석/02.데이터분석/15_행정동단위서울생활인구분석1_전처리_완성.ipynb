### 시도명, 시군구명, 행정동명 추가

#### [참고] 법정동과 행정동          
    * 법정동과 행정동은 우리나라의 행정구역을 구분하는 두 가지 서로 다른 체계입니다.

---

* 법정동     
    * 법률로 지정된 공식 행정구역
    * 1914년 행정구역 통폐합 때 만들어져 현재까지 유지
    * 국토교통부에서 관리
    * 등기부등본, 신분증 등 모든 법적 문서의 주소 표기에 사용
    * 부동산 규제 등 법적 업무의 기준
    * 명칭 변동이 거의 없음

---

* 행정동
    * 행정 편의를 위해 설정된 관리 구역
    * 행정안전부에서 관리
    * 인구수와 행정 수요에 따라 수시로 조정 가능
    * **행정동 코드는 자주 변경되므로, 해당 데이터의 행정동 코드 기준 시점을 확인하여야 함.**
    * **생활인구 데이터는 데이터 일관성 유지를 위해 2016년 집계구, 행정동 코드를 사용하고 있음**
   
---
* 법정동과 행정동의 관계
    * 두 구역은 N:M 관계
    * 하나의 법정동이 여러 행정동으로 나뉠 수 있음
    * 여러 법정동이 하나의 행정동으로 통합될 수 있음
    * 일부 지역은 법정동과 행정동이 일치하기도 함
    * 예를 들어, 서울 중구의 경우 법정동은 87개지만 행정동은 17개만 존재
    * 반대로 송파구 잠실동(법정동)은 잠실본동, 잠실2동, 잠실3동, 잠실7동이라는 4개의 행정동으로 나뉨

# 시각화

**목표**  
- 분석 대상: 4분기(10월~12월) 주말  

- 필터 조건: 서울에 거주하는 여성  

- 목표: 이들이 어느 행정동에 가장 많이 방문했는지 시각화 (지도에 folium으로)

- 4분기(10~12월) 주말에 서울에 사는 여성들이 가장 많이 방문한 행정동

```python
import folium
```

```python
# 1. 서울 + 4분기 + 주말만 필터링
df_filtered = df_all[
    (df_all['시도명'] == '서울') &
    (df_all['월'].isin([10, 11, 12])) &
    (df_all['요일'].isin(['Sat', 'Sun']))  # 요일 수정
].copy()

# 2. 여성 전체 인구 합산
female_columns = [col for col in df_filtered.columns if col.startswith('여자')]
df_filtered['여성총합'] = df_filtered[female_columns].sum(axis=1)

# 3. adm_cd 컬럼 추가
df_filtered['adm_cd'] = df_filtered['행정동코드'].astype(str)  # 꼭 문자열로 맞춰줘야 GeoJSON과 매핑됨

# 4. adm_nm 컬럼 생성
df_filtered['adm_nm'] = '서울특별시 ' + df_filtered['시군구명'] + ' ' + df_filtered['행정동명']

# 5. 그룹핑 (행정동코드 단위로 여성 인구 총합)
df_grouped = df_filtered.groupby(['adm_cd', '행정동명', 'adm_nm'])['여성총합'].sum().reset_index()
```

```python
df_grouped
```

```python
import folium
import json
from folium import Choropleth, GeoJson, GeoJsonTooltip

# 지도 중심은 서울시청
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11, tiles='cartodbpositron')

# geojson 파일 불러오기
geo_path = "C:\\Users\\User\\Documents\\VSCode\\Data School\\data school\\00.data\\geojson\\geojson\\HangJeongDong_ver20161231.geojson"  # 파일명 바꿔줘
with open(geo_path, encoding='utf-8') as f:
    geo = json.load(f)

# 1. Choropleth (초록 계열로 변경: 'YlGn')
Choropleth(
    geo_data=geo,
    data=df_grouped,
    columns=['adm_cd', '여성총합'],
    key_on='feature.properties.adm_cd',
    fill_color='YlGn',   # 🔹초록 계열
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='여성 유동 인구수 (4분기 주말 기준)'
).add_to(m)

# 2. 행정동 이름 고정 표시
GeoJson(
    geo,
    name='행정동 이름',
    style_function=lambda x: {
        'fillOpacity': 0,
        'color': 'black',
        'weight': 0.5
    },
    tooltip=GeoJsonTooltip(
        fields=['adm_nm'],
        aliases=['행정동명:'],
        localize=True,
        sticky=False,
        labels=True,
        style="""
            background-color: white;
            border: 1px solid black;
            border-radius: 3px;
            box-shadow: 3px;
        """,
        max_width=800,
    )
).add_to(m)

# 3. 지도 저장
m.save("서울_여성유동인구_4분기주말.html")
```

### 분석 목표

### 주요 유입 지역 (상위 행정동)

| 지역명             | 주요 특징                              | 유입 유형             |
|------------------|---------------------------------------|--------------------|
| **은평구 진관동**   | 신규 아파트 단지 밀집, 은평뉴타운 중심         | 거주 + 소비 복합형      |
| **은평구 대조동**   | 불광역 인접, 전통시장·상권 활성화           | 교통 + 소형 상권형     |
| **서대문구 신촌동** | 연세대, 이화여대 인접, 대학가 중심 상권        | 대학 + 소비 중심형     |
| **마포구 서교동**   | 홍대입구 중심, 20~30대 유동 인구 집중       | 문화 + 쇼핑 중심형     |
| **영등포구 여의동** | 더현대서울, IFC몰 등 복합 쇼핑몰 위치         | 외부 유입 소비 중심형   |
| **강남구 역삼1동** | 강남역 중심, 직장인·쇼핑·유흥시설 밀집       | 오피스 + 소비 결합형   |

### 유입 지역 유형 분류

| 유형              | 지역 예시              | 공통 특징                         |
|------------------|-----------------------|----------------------------------|
|  **거주+소비 복합형** | 진관동, 대조동             | 주거지 주변 상권 or 교통 요지        |
|  **대학·문화 중심형** | 신촌동, 서교동             | 20~30대 여성, 쇼핑·문화 중심 상권     |
|  **외부 유입 상권형** | 여의동, 역삼1동             | 대형 몰/업무지구 중심 외부 유입 ↑    |

### 결론
- 유동 여성 인구는 **상업·문화·교통이 결합된 지역**에 집중
- 소비 중심 지역(홍대, 강남, 여의도)과  신규 거주지 개발 지역(진관동) 모두 높은 유입
- **20~40대 여성이 많이 이동하는 지역**을 중심으로 마케팅·상권 전략 수립 시 고려 필요
